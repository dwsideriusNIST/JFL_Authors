```python
import json
from urllib.request import urlopen
from pprint import pprint
from copy import deepcopy
```

```python
#makes the cells extarwide when you run this cell, might be useful for longer lists
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
adsorbentsurl = "http://adsorbents.nist.gov/api/materials.json"
webrequest = urlopen(adsorbentsurl)
loadwebrequest = json.load(webrequest)
pprint(loadwebrequest)
```

```python
#splits the adsorbents into MOFs, Zeolites, Carbon, or other. The first three groups are fairly large so that's why.

names = [[y for (x, y) in z.items() if x == "Name" or x == "Synonyms"] for z in loadwebrequest]
MOFlist = []
Zeolitelist = []
Carbonlist = []
Otherslist = []
for x in names:
    if "MOF" in x[0]:
        MOFlist.append(x)
    elif "Zeolite" in x[0]:
        Zeolitelist.append(x)
    elif "Carbon" in x[0]: 
        Carbonlist.append(x)
    else: 
        Otherslist.append(x)
pprint(Otherslist)
```

```python
comparisonnames = deepcopy(names)
MOFcomparisonlist = []

for x in range(len(MOFlist)):
    for y in range(len(names)):
        if MOFlist[x][0] in comparisonnames[y][0] and abs(len(MOFlist[x][0]) - len(comparisonnames[y][0])) < int(.6*len(MOFlist[x][0])) and MOFlist[x][0] != comparisonnames[y][0]:
            MOFcomparisonlist.append(MOFlist[x] + comparisonnames[y])

pprint(MOFcomparisonlist)
```
